```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.DataFrame.merge(ride_data, city_data, on = "city" , how = "left")

# Display the data table for preview
df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Total number of rides per city (x-axis)
city_rides = df["city"].value_counts()
city_rides = city_rides.to_frame(name = "city_rides")
city_rides.reset_index(inplace = True)
city_rides.rename(columns = {"index" : "city"}, inplace = True)
df2 = pd.merge(df, city_rides, on = "city", how = "left")

# Average fare (y-axis)
avg_fare = df2.groupby(["city"])["fare"].mean()
avg_fare = avg_fare.to_frame(name = "avg_fare")
avg_fare.reset_index(inplace = True)
avg_fare.rename(columns = {"index" : "city"}, inplace = True)
df2 = pd.merge(df2, avg_fare, on = "city", how = "left")

cities = df2.groupby(["type"])

urban_cities = cities.get_group("Urban")

suburban_cities = cities.get_group("Suburban")

rural_cities = cities.get_group("Rural")

# Build the scatter plots for each city types
plt.scatter(urban_cities["city_rides"],
            urban_cities["avg_fare"],
            urban_cities["driver_count"]*10,
            edgecolors="black",
            facecolors="coral",
            alpha=0.5,
            label="Urban")

plt.scatter(suburban_cities["city_rides"],
            suburban_cities["avg_fare"],
            suburban_cities["driver_count"]*10,
            edgecolors="black",
            facecolors="skyblue",
            alpha=0.5,
            label="Suburban")

plt.scatter(rural_cities["city_rides"],
            rural_cities["avg_fare"],
            rural_cities["driver_count"]*10,
            edgecolors="black",
            facecolors="gold",
            alpha=0.5,
            label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)", fontweight = "bold")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid(b=True,which="major",axis="both",color="lightgrey")

# Create a legend
legend = plt.legend(title = "City Types", loc="best")

# Setting a fixed size for points in legend
legend.legendHandles[0]._sizes = [35]
legend.legendHandles[1]._sizes = [35]
legend.legendHandles[2]._sizes = [35]

# Incorporate a text label regarding circle size
plt.text(43,35,"Note:\nCircle size correlates with driver count per city.")

# Save figure w/o cutting off the note
plt.savefig("../Pyber/01_ride_sharing_data.png", bbox_inches = "tight")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares1 = df2.groupby(["type"])["fare"].sum()
total_fares2 = total_fares1.sum()
total_fares = total_fares1 / total_fares2

# Build Pie Chart
plt.pie(total_fares,
        shadow = True,
        startangle = 150,
        labels = ["Rural","Suburban","Urban"],
        explode=[0,0,0.075],
        colors=["gold","skyblue","coral"],
        autopct="%1.1f%%")

plt.title("% of Total Fares by City Type", fontweight = "bold")

# Save Figure
plt.savefig("../Pyber/02_total_fares.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = df2.groupby(["type"]).count()["ride_id"]
total_rides = total_rides / total_rides.sum()

# Build Pie Chart
plt.pie(total_rides,
        shadow = True,
        startangle = 150,
        labels = ["Rural","Suburban","Urban"],
        explode=[0,0,0.075],
        colors=["gold","skyblue","coral"],
        autopct="%1.1f%%")

plt.title("% of Total Rides by City Type", fontweight = "bold")

# Save Figure
plt.savefig("../Pyber/03_total_rides.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_rides = df2.drop_duplicates(["city"])

total_drivers = city_rides.groupby(["type"])["driver_count"].sum()

total_drivers = total_drivers / total_drivers.sum()

# Build Pie Chart
plt.pie(total_drivers,
        shadow = True,
        startangle = 150,
        labels = ["Rural","Suburban","Urban"],
        explode=[0,0,0.075],
        colors=["gold","skyblue","coral"],
        autopct="%1.1f%%")

plt.title("% of Total Drivers by City Type", fontweight = "bold")

# Save Figure
plt.savefig("../Pyber/04_total_drivers.png")
plt.show()
```

## Analysis

* In general, the more densely populated a city is, the lower the average fare is. This may be because there are more drivers in cities that are urban when compared to suburban or rural areas. The competition from the higher supply of drivers or higher availability of alternative transport options in urban cities may be influencing this trend.

* Rural and suburban cities have disproportionately fewer drivers when comparing to the percentage of rides these types of cities are responsible for. For example, only 2.6% of drivers are based in rural cities, but rides in rural cities make up more than twice that amount of the total rides taken at 5.3%.

* When you total the fares of all the urban cities, they account for the highest percent of revenue by city type at 62.7%. However, it takes proportionally a larger number of drivers to obtain this result (80.9%) than in the rural and suburban cities.
